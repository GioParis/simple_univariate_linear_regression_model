```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

%matplotlib inline

#stock both datasets in a pandas dataframe
database_train = pd.read_csv('train_lr.csv')
database_test = pd.read_csv('test_lr.csv')

#clean-up your dataset
database_train_ok = database_train.dropna()
database_test_ok = database_test.dropna()

#Show up first 5 lignes of your train dataset
database_train_ok.head()
```

```python
#Slice your dataframe and put the feature you need for your train and test dataset
database_x_train = database_train_ok.as_matrix(['x'])
database_y_train = database_train_ok.as_matrix(['y'])

database_x_test = database_test_ok.as_matrix(['x'])
database_y_test = database_test_ok.as_matrix(['y'])

#print one variable to be sure that the slicing works correctly
print (database_y_train)
```

```python
#initialize your plot axes and grid
axes = plt.axes()
axes.grid()
#fit your features with train dataset variables
plt.scatter(database_x_train,database_y_train)
#name each item
plt.title('Plot Example')
plt.ylabel('Y')
plt.xlabel('X')
#Show the plot
plt.show()
```

```python
#initialize univariate linear regression model
regr = linear_model.LinearRegression()
```

```python
#train your algorithm with train dataset variables
regr.fit(database_x_train,database_y_train)
```

```python
#predict dependent variable by using predictor variable
pred_db_y = regr.predict(database_x_train)

#Let's take 50 first values in order to have the same variable size for plotting them
pred_y_ok = pred_db_y[:50]
database_y_test_ok = database_y_test[:50]
database_x_train_ok = database_x_train[:50]
```

```python
#name each item on your scatter plot
plt.title('Y values in test vs Predicted values')
plt.ylabel('Test set')
plt.xlabel('Predicted values')
#feed up your scatter plot and your vector
plt.scatter(pred_y_ok,database_y_test_ok, color='black')
plt.plot(database_x_train_ok,pred_y_ok,color='red')
#show up the univariate linear regression model sample
plt.show()

#print out the intercept and coefficient
print ('The intercept of the model is: {}'.format(regr.intercept_))
print ('The linear regression coefficient is: {}'.format(regr.coef_))
```
